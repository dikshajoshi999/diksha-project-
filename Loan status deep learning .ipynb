# problem statement: to check loan eligibility of customer deep learning model

```python
import pandas as pd
```

```python
A = pd.read_csv("C:/Users/Admin/Downloads/training_set.csv")
```

```python
A = A.drop(labels=["Loan_ID"],axis=1)
```

```python
from ml_txt import replacer
replacer(A)
```

```python
Y = A[["Loan_Status"]]
X = A.drop(labels=["Loan_Status"],axis=1)
```

```python
from ml_txt import data_prep
Xnew=data_prep(X)
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Ynew = pd.DataFrame(le.fit_transform(Y))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Ynew,test_size=0.2,random_state=21)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping

nn = Sequential()
nn.add(Dense(200,input_dim=(Xnew.shape[1])))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(1,activation="sigmoid"))
```

```python
nn.compile(loss="binary_crossentropy",metrics = "accuracy")
nn.fit(xtrain,ytrain,epochs=100,validation_split=0.2,batch_size=4)
```

```python
B = pd.read_csv("C:/Users/Admin/Downloads/testing_set.csv")
```

```python
B = B.drop(labels=["Loan_ID"],axis=1)
```

```python
from ml_txt import replacer,data_prep
replacer(B)
test=data_prep(B)
```

```python
Xnew.shape
```

```python
Q = []
for i in nn.predict(test):
    if(i < 0.5):
        Q.append(0)
    else:
        Q.append(1)
```

```python
predicted_loan_status=le.inverse_transform(Q)
```
