# Short term forecasting example:

```python
import pandas as pd
import numpy as np
```

```python
A = pd.read_csv("C:/Users/Admin/Downloads/Google_Stock_Price_Train.csv")
```

```python
A
```

```python
trd = A[["Open"]].values
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
trd=mm.fit_transform(trd)
```

```python
trd.shape
```

```python
xtrain = []
ytrain = []
for i in range(50,1258,1):
    xtrain.append(trd[i-50:i])
    ytrain.append(trd[i:i+1])
```

```python
xtr = np.array(xtrain)
ytr = np.array(ytrain)
```

```python
xtr.shape
```

```python
ytr.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
nn = Sequential()
nn.add(LSTM(50,return_sequences=True,input_shape=(50,1)))
nn.add(LSTM(100,return_sequences=True))
nn.add(Dense(1))
```

```python
nn.compile(optimizer="adam",loss="mean_squared_error")
nn.fit(xtr,ytr,epochs=10)
```

```python
tsd = pd.read_csv("C:/Users/Admin/Downloads/Google_Stock_Price_Test.csv")
```

```python
tsd.isna().sum()
```

```python
Q = pd.concat([A,tsd])
Q.index = range(0,1278)
```

```python
Q = Q[["Open"]].values
```

```python
xtest=[]
for i in range(1258,1277,1):
    xtest.append(Q[i-50:i])
```

```python
xts = np.array(xtest)
```

```python
xts.shape
```

```python
d = nn.predict(xts[18:19])
```

```python
d = d.reshape(50,1)
```

```python
mm.inverse_transform(d)
```
