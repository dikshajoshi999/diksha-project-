# Problem stmt: To predict sales price of each house:(Deep Learning)

```python
#Read Dataset
```

```python
import pandas as pd                      
```

```python
A = pd.read_csv("C:/Users/Admin/Downloads/training_set (1).csv")
```

```python
#Drop unnecessary columns
```

```python
A = A.drop(labels=["Id"],axis=1)
```

```python
#Missing data treatment
```

```python
from ml_txt import replacer
replacer(A)
```

```python
Y = A[["SalePrice"]]
X = A.drop(labels=["SalePrice"],axis=1)
```

```python
#Preprocessing of data
```

```python
from ml_txt import data_prep
Xnew = data_prep(X)
```

```python
Xnew.shape
```

```python
#Divide data in train and testing set and create DL model
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping

nn = Sequential()
es = EarlyStopping(patience=20)
nn.add(Dense(200,input_dim=(Xnew.shape[1])))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(1,activation="relu"))
```

```python
nn.compile(loss="mean_squared_error")
nn.fit(xtrain,ytrain,epochs=50,validation_split=0.2,batch_size=4,callbacks=es)
```
