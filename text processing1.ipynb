# Text Processing (To check about Restaurant Review whether liked /disliked)

```python
import pandas as pd
```

```python
A=pd.read_csv("C:/Users/Admin/Downloads/Restaurant_Reviews.tsv",sep='\t')
```

```python
A.head()
```

```python
#To remove special characters after each word
Q=[]
from re import sub
for i in A.Review:
    Q.append(sub("[^A-Za-z0-9 ]","",i.upper()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
words_vect = cv.fit_transform(Q).toarray()
```

```python
X = words_vect
Y = A.Liked
```

```python
X.shape
```

```python
#Divide Data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=21)
```

```python
#Create DL model
from keras.models import Sequential
from keras.layers import Dense,Dropout

nn=Sequential()
nn.add(Dense(1000,input_dim = (2067)))
nn.add(Dropout(0.1))
nn.add(Dense(1000))
nn.add(Dropout(0.1))
nn.add(Dense(1000))
nn.add(Dropout(0.1))
nn.add(Dense(1,activation="sigmoid"))

```

```python
nn.compile(loss="binary_crossentropy",metrics="accuracy")
nn.fit(xtrain,ytrain,epochs=10)
```

```python
#Calculate predictions:
W=[]
for i in nn.predict(xtest):
    if(i[0] < 0.5):
        W.append(0)
    else:
        W.append(1)
```

```python
W
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,W)
```

```python
#To manually enter input and see if food is liked or not
Rev = input()
rev1 = cv.transform([Rev.upper()]).toarray()
for i in nn.predict(rev1):
    if(i[0] < 0.5):
        print("Not liked")
    else:
        print("liked")
```

```python
#Performing Stemming and Lemmitization:
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
from re import sub
Q = []
for i in A.Review:
    W = sub("[^A-Za-z0-9 ]","",i.upper())
    Q1 = []
    for j in W.split(" "):
        Q1.append(wnl.lemmatize(ps.stem(j)))
        
    Q.append(Q1)
        
```
